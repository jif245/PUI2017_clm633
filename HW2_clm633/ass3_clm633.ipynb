Repeat the steps above 1-7 with a CSV file that contains _a date/time column and a numerical value_. Plot the numerical value against the date/time (hint: make sure your tick labels are readable. you can use they keyword rot in the df.plot() method, as for example df.plot(....., rot=90) to rotate the tick labels by 90 degrees.

```python
df2 = pd.read_csv(ds2)
df2.head()
```

```python
to_plot = df2[["DATE", "NUMBER OF PERSONS INJURED"]].groupby("DATE").sum().reset_index()
to_plot["DATE"] = pd.to_datetime(to_plot["DATE"])
```

```python
import seaborn as sns
ax = to_plot.plot("DATE", "NUMBER OF PERSONS INJURED")
ax.set_title("number of people injured in accidents (presumably) on this day")
```

### Caption
Collision report data over a number of years, looks like. X axis is time, Y axis is # of people injured aggregated by day.

```python
from pandas.tools.plotting import autocorrelation_plot
ax = autocorrelation_plot(to_plot.set_index("DATE")["NUMBER OF PERSONS INJURED"])
ax.set_ylim([-.25, .25])
ax.set_xlim([0, 400])
ax.set_xlabel("Lag (days)")
ax.set_title("Autocorrelation plot of # injuries per day")
```
